```python
import os
import sys
import numpy as np
import json
import matplotlib.pyplot as plt
```

```python
#os.listdir()
#Model type results
#MTR=os.path.join("..","..","Results","VAE","Basic_CNN_EDM")
MTR=os.path.join("..","..","Results","VAE","PCNN_EDM")

Res_dir=os.path.join(MTR,"Test_7")
Losses=os.path.join(Res_dir,"loss_results.npy")

m_config=json.load(open(os.path.join(Res_dir,"config.json")))
```

```python
#m_config
list(m_config["model"]["model_params"]["losses_weigths"].keys())
```

```python
model_losses=list(m_config["model"]["model_params"]["losses_weigths"].keys())

figure,axs=plt.subplots(1,1)

train_losses=np.array(np.load(Losses,allow_pickle=True).tolist()["train"][model_losses[0]])
test_losses=np.array(np.load(Losses,allow_pickle=True).tolist()["test"][model_losses[0]])
#axs.plot(np.log(train_losses[-1000:]))
#axs.plot(np.log(test_losses[-1000:]))
axs.plot((train_losses[:]))
axs.plot((test_losses[:]))
```

```python
figure,axs=plt.subplots(1,1)

for t in os.listdir(MTR):
    t_res=os.path.join(MTR,t)
    Losses=os.path.join(t_res,"loss_results.npy")
    m_config=json.load(open(os.path.join(Res_dir,"config.json")))
    model_losses=list(m_config["model"]["model_params"]["losses_weigths"].keys())

    try:
        train_losses=np.array(np.load(Losses,allow_pickle=True).tolist()["train"][model_losses[0]])
        test_losses=np.array(np.load(Losses,allow_pickle=True).tolist()["test"][model_losses[0]])
        axs.plot((train_losses[:]),'r')
        axs.plot((test_losses[:]),'b')
    except:
        print(t)
```
